```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df_temp = pd.read_csv('dist/rank_TOP_1_California.csv')

top_5_C = ['California','Texas','Florida','New York','Illinois']

rank = 1
for i in top_5_C:
    file_name = "rank_TOP_" + str(rank) + "_" + i + ".csv"
    path = 'dist/' + file_name
    rank += 1
    df_temp = pd.read_csv(path)
    unique_date = df_temp['Date'].unique()
    sum_confirm_by_date = df_temp.groupby('Date')['Total Confirmed'].sum()
    new_df = pd.DataFrame({'Date': unique_date, 'Total Confirmed': sum_confirm_by_date})
    plt.rcParams.update({'font.size':8})
    plt.figure(figsize=(10, 7))
    plt.plot(new_df['Date'], new_df['Total Confirmed'], color='red', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Total Confirmed')
    plt.title('Rates of Confirmed and Date in ' + i)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
